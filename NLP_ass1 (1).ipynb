```python
from google.colab import drive
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re
```

```python
file_path = '/content/drive/MyDrive/wikipidedia messi html.txt'  # Replace with the actual file name
with open(file_path, 'r') as file:
    html_content = file.read()

# Now, you can use the 'dataset_content' variable as your text dataset
print(html_content)
```

```python

soup = BeautifulSoup(html_content, 'html.parser')

# Extract text content without HTML tags
text_content = soup.get_text(separator=' ')

# Remove extra whitespaces
text_content = re.sub('\s+', ' ', text_content).strip()

print(text_content)
```

```python

output_file_path = '/content/drive/MyDrive/text content.txt'

with open(output_file_path, 'w') as output_file:
    for word in text_content:
        output_file.write(word)

print(f"Unique words saved to: {output_file_path}")
```

```python
nltk.download('punkt')
nltk.download('stopwords')

text_content = text_content.lower()

# Remove special characters and digits
text_content = re.sub(r'[^a-zA-Z\s]', '', text_content)

tokenized_words = word_tokenize(text_content)

stop_words = set(stopwords.words('english'))
tokenized_words = [word for word in tokenized_words if word not in stop_words]

porter = PorterStemmer()
words = [porter.stem(word) for word in tokenized_words]

print("words after preprocessing :",words)

unique_words= set(words)

print("Unique words :", unique_words)
```

```python
unique_words_len_3_or_less = [word for word in unique_words if len(word) <= 3]
print(unique_words_len_3_or_less)
```
